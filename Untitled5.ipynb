# problem 2 10-3

```r
student<-read.table("http://web.stanford.edu/~hastie/CASI_files/DATA/student_score.txt" , header =TRUE )
A<-cor(student);A
```

```r
theta_hat <- function(data, indices) {
  d <- data[indices,] #allows boot to select sample
  A<-cor(d)
  ev<-eigen(A)
  return(max(ev$values)/sum(ev$values))
}

reps <- boot(data=student, statistic=theta_hat, R=2000);reps
```

```r
hist(reps$t,col = '#00AFBB',breaks = 20)
abline(v=0.69, lwd=3)
```

# problem3   10-6

```r
Boot_nonparam <- function(x, B, fun){
  x <- as.matrix(x)
  n <- nrow(x)
  f0 <- fun(x)  # get size of output
  fmat <- matrix(0, length(f0), B)

  for (b in 1:B) {
    i <- sample(1:n, n, replace = TRUE)
    fmat[, b] <- fun(x[i,])
  }

  drop(fmat)
}

cor_c <- function(x) {
  return(cor(x[,2],x[,3]))
  }

set.seed(123)
B <- 2000

a<-read.csv("kidney",header=TRUE)#157개 데이터
data <- as.matrix(a,ncol=2)
result <- Boot_nonparam(data, B, cor_c)
print(round(sd(result),3))
```

# problem 5

```r
data<-runif(n=50,min=0,max=1)
data
```

```r
theta_hat<-data.frame(rep(1,2000),rep(1,2000))
#parametric bootstrap
for (i in 1:2000){
  theta_hat[i,2]<-max(runif(n=50,min=0,max=1))
}
colnames(theta_hat)<-c('non-param','param')
```

```r
#nonparametric bootstrap
for (i in 1:2000){
  theta_hat[i,1]<-max(sample(data,50,replace=TRUE))
}
theta_hat
```

```r
real_dist<-function(x){
  50*x^(49)
}
```

```r
plot(hist(theta_hat[,1],breaks='scott',main = 'non-parametric-bootstrap' ),col=rgb(0,0,1,1/4), xlim=c(0,10),add=TRUE)
plot(hist(theta_hat[,2],breaks='scott',main = 'parametric-bootstrap'    )  , col=rgb(1,0,0,1/4), xlim=c(0,10), add=TRUE)
plot(seq(0,1,by=0.01),real_dist(seq(0,1,by=0.01)),col=rgb(0,0.8,1),type='l', main='true-distribution')
legend("topleft", legend=c("true-distribution"), col=rgb(0,0.8,1), lty=1:2)
```
